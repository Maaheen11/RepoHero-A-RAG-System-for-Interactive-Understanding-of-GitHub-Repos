## More Example test cases

### FastAPI — `test_fastapi.csv`

**1. Question**

> What is FastAPI and what language is it written in?

**Expected answer**

FastAPI is a modern web framework used for building APIs using Python. It is designed for high performance and is built on top of Starlette for the web layer and Pydantic for data validation. It supports asynchronous programming and automatic OpenAPI documentation generation.

**2. Question**

> What does the @app.get decorator do in FastAPI?

**Expected answer**

The @app.get decorator registers a function as a handler for HTTP GET requests at a specific path. When a request arrives at that route, FastAPI executes the decorated function. The function’s return value is automatically converted into a JSON response.

**3. Question**

> What role does type annotation play in FastAPI endpoint functions?

**Expected answer**

Type annotations allow FastAPI to validate request inputs automatically. They also help generate API documentation and ensure proper data conversion. The framework relies heavily on Python type hints for request validation.

**4. Question**

> How does FastAPI generate API documentation automatically?

**Expected answer**

FastAPI analyzes route definitions, parameters, and Pydantic models to build an OpenAPI schema. This schema is used to generate interactive documentation interfaces. Developers can explore and test the API through Swagger UI or ReDoc.

**5. Question**

> How are routers used to organize endpoints in FastAPI repositories?

**Expected answer**

Routers allow endpoints to be grouped into separate modules using APIRouter. Each router can contain related routes and be included in the main application. This improves code organization and scalability.

### Requests — `test_requests.csv`

**1. Question**

> How was a simple HTTP GET request sent to a URL?

**Expected answer**

A GET request can be sent by calling the get function with the target URL which internally constructs and dispatches a request and returns a response object.

**2. Question**

> How was JSON data transmitted in an HTTP request?

**Expected answer**

JSON data can be sent by passing a dictionary to the json parameter which automatically serializes the object and sets the application/json header.

**3. Question**

> How was persistent connection reuse achieved across multiple requests?

**Expected answer**

Connection reuse is achieved by creating a session object which manages connection pooling and persistent settings.

**4. Question**

> How were network related errors represented during requests?

**Expected answer**

Network errors are raised as exceptions derived from RequestException such as ConnectionError or Timeout.

**5. Question**

> How were large responses streamed instead of downloaded entirely?

**Expected answer**

Streaming can be enabled by passing stream=True and iterating over response.iter_content to process chunks incrementally.

### Pydantic — `test_pydantic.csv`

**1. Question**

> What does the model_copy() method do on a BaseModel instance, and what do the update and deep parameters control?

**Expected answer**

model_copy() returns a new instance of the model with optionally overridden field values. The update parameter accepts a dictionary of field names to new values that are applied to the copy. The deep parameter, when set to True, performs a deep copy of all field values so that nested mutable objects are not shared between the original and the copy.

**2. Question**

> What does ConfigDict(populate_by_name=True) enable on a model that uses field aliases?

**Expected answer**

ConfigDict(populate_by_name=True) allows a model field to be set using either its Python attribute name or its alias during validation. Without this setting, a field with an alias can only be populated using the alias name, not the original Python field name.

**3. Question**

> What does the strict parameter do when passed to model_validate() or TypeAdapter.validate_python()?

**Expected answer**

When strict=True is passed, Pydantic disables all type coercion and requires the input value to already be exactly the declared type. For example, passing the string "123" for an int field will raise a ValidationError in strict mode whereas it would normally be coerced to the integer 123.

**4. Question**

> How does Pydantic's model_dump() interact with @computed_field when include or exclude parameters are passed?

**Expected answer**

When include or exclude sets are passed to model_dump(), Pydantic filters the output fields accordingly, and @computed_field properties are subject to the same filtering — they are included or excluded based on whether their property name appears in the include or exclude set. If a computed field is excluded, its property is never called during serialization.

**5. Question**

> What is the overall purpose of the Pydantic library and what problem does it solve?

**Expected answer**

Pydantic is a Python data validation library that uses type annotations to define the shape and constraints of data, automatically validating and coercing input values when model instances are created. It solves the problem of safely parsing untrusted external data such as API payloads, configuration files, and database records into strongly-typed Python objects with clear error reporting when data does not conform to the expected schema.
